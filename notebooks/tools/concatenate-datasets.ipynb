```python
# read and split into train and test sets
import numpy as np
np.random.seed(42)
IN_SENTENCES = "/home/railway/Documents/diploma/master-thesis/data/semeval/es/es-sentences-all.txt"
IN_LABELS = "/home/railway/Documents/diploma/master-thesis/data/semeval/es/es-labels-all.txt"
OUT_TRAIN_SENTENCES = "/home/railway/Documents/diploma/master-thesis/data/semeval/es/es-train-sentences.txt"
OUT_TRAIN_LABELS = "/home/railway/Documents/diploma/master-thesis/data/semeval/es/es-train-labels.txt"
OUT_TEST_SENTENCES = "/home/railway/Documents/diploma/master-thesis/data/semeval/es/es-test-sentences.txt"
OUT_TEST_LABELS = "/home/railway/Documents/diploma/master-thesis/data/semeval/es/es-test-labels.txt"
OUT_PATHS = [OUT_TEST_SENTENCES, OUT_TEST_LABELS]
SPLIT_RATIO = 0.8

with open(IN_SENTENCES) as in_sent_f, open(IN_LABELS) as in_labs_f:
    sentences = in_sent_f.readlines()
    labs = in_labs_f.readlines()
    assert len(sentences) == len(labs)
    sentences = np.array(sentences)
    labs = np.array(labs)
    print(f"sentences shape: {sentences.shape}, labs shape {labs.shape}")
    print(f"sentences[1684]: {sentences[1684]} - lab: {labs[1684]}")
    print(f"sentences[65]: {sentences[65]} - lab: {labs[65]}")
    indices = np.arange(len(sentences))
    np.random.shuffle(indices)
    print(f"indices: {indices}")
    sentences = sentences[indices]
    labs = labs[indices]
    print(f"sentences[0]: {sentences[0]} - lab: {labs[0]}")
    print(f"sentences[0]: {sentences[1]} - lab: {labs[1]}")
    splt_ind = int(SPLIT_RATIO * len(indices))
    print(f"splt_ind {splt_ind}")
    train_sentenses = sentences[:splt_ind]
    train_labs = labs[:splt_ind]
    test_sentenses = sentences[splt_ind:]
    test_labs = labs[splt_ind:]
#     data_list = [train_sentenses, train_labs, test_sentenses, test_labs]
    data_list = [test_sentenses, test_labs]
    assert len(train_labs) == len(train_sentenses)
    assert len(test_labs) == len(test_sentenses)
    print(f"train len: {len(train_labs)} - test len: {len(test_labs)}")
    for data, out_path in zip(data_list, OUT_PATHS):
        with open(out_path, "w") as out_f:
            for sample in data:
                out_f.write(sample)
    
```
